```python
import numpy as np
import time

tic = time.time()
G = 6.67e5
threshold = 0.1
ts = 1e-4                                                 #time step
M = np.load("../q2_input/masses.npy")       #masses
X = np.load("../q2_input/positions.npy")     #postions
V = np.load("../q2_input/velocities.npy")    #velocities
A = np.zeros((100,100))
chk = np.ones((100,100))

def calcAccl():
    global A
    global chk
#     global X
    
    D = X.reshape(1, 100, 2)
    D = np.repeat(D, 100, axis=0)

    temp = np.repeat(X, 100, axis=0)
    temp = temp.reshape(100,100,2)
    D = D - temp
    temp = D
    
#     print(M)
    chk = np.linalg.norm(D, axis=2)
    np.fill_diagonal(chk, 1)
    D = np.power(chk, 3)

#     D = (D * M.reshape(1,100)).reshape(100, 100, 1)
    D = D.reshape(100,100,1)
#     print(np.repeat(M.reshape(1,100,1), 2, axis=2))
    temp = temp * np.repeat(M.reshape(1,100,1), 2, axis=2)
    D = G * (temp / np.repeat(D, 2, axis=2))
    D = np.sum(D, axis=1)

    A = D

iter = 0
while True:
    val = chk.min()
    print(val)
    if val < threshold:
        #print to files
        np.save('finalPosCmp.npy', X, allow_pickle = False)
        np.save('finalVelCmp.npy', V, allow_pickle = False)
        print("iter = %d" % iter)
        break
    calcAccl()
    X = X + V * ts + 0.5 * A * ts * ts
    V = V + A * ts
    iter += 1
        
print("--- %s seconds ---" % (time.time() - tic))
```
